```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
response = requests.get("https://myanimelist.net/topanime.php?limit={}")

data_rank = []
data_title = []
data_episode = []
data_dates = []
data_members = []
data_score = []
for index in range(0, 451, 50):
    page = requests.get(f'https://myanimelist.net/topanime.php?limit={index}')
    soup = BeautifulSoup(response.content, "html.parser")
    anime_list = soup.find_all("tr", class_="ranking-list")
    for anime in anime_list:
        rank = anime.find("span", class_="top-anime-rank-text").text.strip()
        title = anime.find("div", class_="di-ib clearfix").a.text.strip()
        total = anime.find("div", class_="information di-ib mt4").text.strip()
        clean_info = total.replace('<br>', '\n').replace('<div class="information di-ib mt4">', '').replace('</div>', '').strip()
        lines = clean_info.split('\n')
        total_episode = lines[0].strip()
        dates = lines[1].strip()
        members = lines[2].strip()
        score_element = anime.find("span", class_="text on score-label score-9")
        score_element1 = anime.find("span", class_="text on score-label score-8")
        score = score_element.text.strip() if score_element else score_element1.text.strip()
        
        data_rank.append(rank)
        data_title.append(title)
        data_episode.append(total_episode)
        data_dates.append(dates)
        data_members.append(members)
        data_score.append(score)
```

```python
data = {
        'Ranking' :data_rank,
        'Title' :data_title,
        'Total Episode' :data_episode,
        'Dates' :data_dates,
        'Total Member' :data_members,
        'Score' :data_score,

            
}
pd.DataFrame(data).to_excel('animelist.xlsx')
```
